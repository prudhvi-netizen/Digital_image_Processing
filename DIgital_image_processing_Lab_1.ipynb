```python
import cv2
from matplotlib import pyplot as plt

# Specify the path to your image
image_path = 'Prudhvi.jpg'  # Replace with the actual path to your image file

# Read the image
image = cv2.imread(image_path)

# Convert BGR image to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
#coordinates for color image
image_rgb.shape
```

```python
#intensity for original image
image_rgb
```

```python
#converting original image to grayscale
gray_img = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
plt.imshow(gray_img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```
